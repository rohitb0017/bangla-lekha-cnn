```python
from google.colab import files
import os
import shutil

# Define the target directory
target_dir = '/content/banglalekha_dataset/BanglaLekha_Dataset'
#os.makedirs(target_dir, exist_ok=True)  # Create the directory if it doesn't exist

# Upload multiple files
uploaded_files = files.upload()

# Move each uploaded file to the target directory
for filename in uploaded_files.keys():
    shutil.move(filename, os.path.join(target_dir, filename))

# List uploaded files in the target directory
print("Uploaded files:", os.listdir(target_dir))
```
